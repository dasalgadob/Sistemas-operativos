```
import numpy as np
import matplotlib.pyplot as plt

posix_time = (3.49, 1.85, 0.98, 0.60, 0.49)
mp_time = (1.32, 0.74, 0.44, 0.29, 0.30)
ind = np.arange(len(posix_time))  # the x locations for the groups
width = 0.4  # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, posix_time, width,
                color='SkyBlue', label='POSIX')
rects2 = ax.bar(ind + width/2, mp_time, width,
                color='IndianRed', label='OPENMP')
ax.set_ylabel('Tiempo (s)', fontsize=16)
ax.set_xlabel('Hilos ', fontsize=16)
ax.set_title('Tiempo de respuesta efecto borroso resolución 720', fontsize=18)
ax.set_xticks(ind)
ax.set_xticklabels(('1', '2', '4', '8', '16'))
ax.legend()

def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    xpos = xpos.lower()  # normalize the case of the parameter
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0.5, 'right': 0.57, 'left': 0.43}  # x_txt = x + w*off

    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()*offset[xpos], 1.01*height,
                '{}'.format(height), ha=ha[xpos], va='bottom')


autolabel(rects1, "left")
autolabel(rects2, "right")
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2,0,4))
plt.show()
```
